```python
import pandas as pd
import numpy as np

#read in 2018 CLUMPS results
clumps_results = pd.read_csv('CLUMPS_2018_results.txt',sep='\t')
#Select results with P< 0.5
clumps_results = clumps_results.loc[clumps_results['CLUMPS_P']<0.5]

uniprot_file = '/mnt/nfs/ro_disks/canine-c0820e9f17cde673ca995479dc35c9ae/uniprot_map/huniprot2pdb.run18.filt.txt'
uniprotmap = pd.read_csv(uniprot_file,
                        names=['u1', #uniprot id
                         'u2', #also uniprot iD (should be the same for AF, )
                         'struct', #name of structure/chain - 1 chain per peptide in AF?
                         'blat', #- not relevant for AF
                         'AA:AA list' #colon separated amino acid positions - should be 1:1, 2:2 etc. for AF? - revisit this
                        ],
                        sep='\t')
```

```python
mc3maf = pd.read_csv('MC3-14k-merge_v1.variant_classification.maf',sep='\t')
mc3maf
```

```python
#select genes of interest

#previously identified genes
goi = [
    'ATM',
    'RHOA',
    'CNOT9',
    'RQCD1',
    'DDX3X',
    'MTOR'
    ]

#random genes
np.random.seed(1234)

#remove zinc fingers - add a lot of redundant structures to permute over
clumps_results = clumps_results.loc[clumps_results['GENE_NAMES'].notna()]
clumps_results = clumps_results.loc[~clumps_results['GENE_NAMES'].str.contains('ZNF')]
clumps_results
```

```python
random_goi = np.random.choice(clumps_results['GENE_NAMES'],50,replace=False)
goi = goi + list(random_goi)
goi
```

```python
mc3maf = mc3maf.loc[mc3maf['Hugo_Symbol'].isin(goi)]
mc3maf
```

```python
clumps_results = clumps_results.loc[clumps_results['GENE_NAMES'].isin(goi)]
clumps_results
```

```python
uniprotmap = uniprotmap.loc[uniprotmap['struct'].isin(clumps_results['PDBID-CHAIN'])]
uniprotmap = uniprotmap.loc[uniprotmap['u1'].isin(clumps_results['UNIPROT_ID'])]
uniprotmap.reset_index()
uniprotmap
```

```python
import os

out_dir = 'uniprot_chunks'
chunk_length= 10
if not os.path.exists(out_dir):
    os.mkdir(out_dir)
gl = len(uniprotmap.index)
start = 0
end = chunk_length
i = 0
while end < gl:
    if i < 10:
        num = '0000%s' %i
    elif i < 100:
        num = '000%s' %i
    elif i < 1000:
        num = '00%s' %i
    fn = 'uniprot_chunks/uniprot50genes_%s.gz' %num
    uniprotmap.iloc[start:end].to_csv(fn,
            sep='\t',
            header=False,
            index=False)
    i = i + 1
    start = end
    end = end+chunk_length
end = gl

if i < 10:
    num = '0000%s' %i
elif i < 100:
    num = '000%s' %i
elif i < 1000:
    num = '00%s' %i
fn = 'uniprot_chunks/uniprot50genes_%s.gz' %num
uniprotmap.iloc[start:end].to_csv(fn,
        sep='\t',
        header=False,
        index=False)
```
